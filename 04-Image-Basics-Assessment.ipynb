#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
vertices = np.array([[425,400], [250,700], [600,700]])
vertices.shape
```

```python
mat = vertices.reshape(-1,1,2)
print(mat.shape)
print(mat)
print(mat.dtype)
```

```python
cv2.polylines(fix_img,[mat], isClosed=True, color=(0,0,255),thickness=15)
plt.imshow(fix_img)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import numpy as np
import cv2


def new_circle(event,x,y,flags,params):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,color=(0,0,255),thickness=10)



img = cv2.imread(r"C:\Users\admin\Desktop\Python_CV\DATA\dog_backpack.png")


cv2.namedWindow(winname="my_image")
cv2.setMouseCallback("my_image",new_circle)

while True:
    cv2.imshow("my_image",img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
