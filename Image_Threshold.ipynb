```python
img = cv2.imread(r"C:\Users\admin\Desktop\Python_CV\DATA\crossword.jpg",0)
plt.imshow(img,cmap = "gray")
```

```python
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img,200,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1,alpha=0.6,src2=th2,beta=0.4,gamma=0)
show_pic(blended)
```
